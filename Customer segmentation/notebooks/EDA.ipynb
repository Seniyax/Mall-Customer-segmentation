```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder,RobustScaler
from sklearn.cluster import KMeans,AgglomerativeClustering
```

```python
db = pd.read_csv('/content/Mall_Customers.csv')
db.head()
```

```python
db.info()
```

```python
db.describe()
```

```python
db.isna().sum()
```

```python
Age_anual_income_plot = px.scatter(x=db['Age'],
                                   y=db['Annual Income (k$)'])

Age_anual_income_plot.show()
```

```python
plt.figure(1,figsize=(15,6))
sns.countplot(y = 'Gender',data=db)
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(1,figsize=(15,7))
for gender in ['Male','Female']:
  plt.scatter(x='Age',y='Annual Income (k$)',data=db[db['Gender']==gender],s = 200,alpha=0.5,label=gender)
plt.title('Age vs Salary w.r.t Gender')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
plt.figure(1,figsize=(8,6))
sns.scatterplot(data=db,x='Gender',y='Spending Score (1-100)',palette='Set2')
plt.title('Gender vs Spending score')
plt.xlabel('Gender')
plt.ylabel('Spending score(1-100)')
##plt.grid(True)
plt.show()
```
